## Extracting Images

```python
import fitz
import PIL.Image
import io
import numpy as np

pdf = fitz.open("z.pdf")
for page in pdf: # get each page of pdf file
    for imageData in page.get_images(): # get all images in pdf
        image = pdf.extract_image(imageData[0])
        extractedImage = PIL.Image.open(io.BytesIO(image["image"]))
        # extractedImage = PIL.Image.open(io.BytesIO(image["image"])).show() # or write this line instead of above line to show the images
        print(np.array(extractedImage)) # printing the numpy array data which will be the input of LLM
```
